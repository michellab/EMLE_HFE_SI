```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("mnsol_dataset.csv")
df
```

```python
sns.set(style="whitegrid", font_scale=1.2, palette="colorblind")

plt.figure(figsize=(8, 6))
sns.histplot(df["Exp [kcal/mol]"], bins=6, kde=True, linewidth=2)
plt.xlabel(r"$\Delta G_{vac\rightarrow sol}^{EXP}$ [kcal.mol$^{-1}$]")
plt.ylabel("# Molecules")
plt.xlim(-15, 5)
plt.tight_layout()
plt.savefig("fig_si_experimental_dG_mnsol.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
